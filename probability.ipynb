# Cumulative distribution function (CFD)
**Cumulative Distribution Function (CDF)** in statistics tells us the probability that a random variable ($X$) takes on a value less than or equal to a certain number ($x$). It gives a *cumulative* view of how probabilities accumulate as we move through the possible values of the variable. Mathematically, it is defined as: $$F(x)=P(X\le x)$$
where Here, $F(x)$ denotes the cumulative probability distribution function of $X$.

Example:

Let's consider an example with a discrete random variable that represents the outcome of rolling a fair six-sided die:

* The possible outcomes od $X$ are `[1,2,3,4,5,6]`.
* Since the die is fair, each outcome has an equal probability of `1/6`.

Now, let's define the cumulative distribution function $F(x)$:

* $F(1)=P(X\le 1) = \frac{1}{6}$; because rolling a 1 has a probability of $\frac{1}{6}$.
* $F(2)=P(X\le 2) = \frac{2}{6}$; because rolling a 1 or 2 has a probability of $\frac{1}{6}+\frac{1}{6}=\frac{1}{3}$.
* $F(3)=P(X\le 3) = \frac{3}{6}$; because rolling a 1, 2, or 3 has a probability of $\frac{1}{6}+\frac{1}{6}+\frac{1}{6}=\frac{1}{2}$.
* $F(4)=P(X\le 4) = \frac{4}{6}$; because rolling a 1, 2, 3 or 4 has a probability of $\frac{1}{6}+\frac{1}{6}+\frac{1}{6}+\frac{1}{6}=\frac{2}{3}$.
* $F(5)=P(X\le 5) = \frac{5}{6}$; because rolling a 1, 2, 3, 4 or 5 has a probability of $\frac{1}{6}+\frac{1}{6}+\frac{1}{6}+\frac{1}{6}+\frac{1}{6}=\frac{5}{6}$.
* $F(5)=P(X\le 5) = \frac{6}{6}$; because rolling a 1, 2, 3, 4, 5 or 6 has a probability of $\frac{1}{6}+\frac{1}{6}+\frac{1}{6}+\frac{1}{6}+\frac{1}{6}+\frac{1}{6}=1$.

So, the cumulative distribution function $F(x)$ increases as $x$ increases, because it accumulates the probabilities up to $x$.


## Python code

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Outcome X
X = np.arange(1, 7)
print(f"Possible outcomes are {X}")
```

If you roll a die, the probability of obtaining 1, 2, 3, 4, 5, or 6 is 1/6. The probability density function (PDF) or the probability that you will get exactly 2 will be 0.1667.

```python
unique_X, counts = np.unique(X, return_counts=True) # If True, also return the number of times each unique item appears
# PDF is then defined as
pdf = counts / len(X)
print(f"PDF is {pdf}")
```

```python
plt.bar(unique_X, pdf)
plt.xlabel('X (Outcome)')
plt.ylabel('PDF')
plt.title('Probability Density Function (PDF)');
```

But the cumulative distribution function of 2 is 0.3333

```python
cdf = np.cumsum(pdf) # cumulative sum of the elements
```

```python
plt.step(unique_X, cdf)
plt.xlabel('X (Outcome)')
plt.ylabel('CDF')
plt.title('Probability Density Function (PDF)');
```

## For *n* 6-sided dice

Now, for example, for n 6-sided dice?

```python
# Let's create a function for n 6-sided dice
def prob(num_rolls, n):
  # Generate random outcomes for each pair of dice
  outcomes = np.random.randint(1, 7, (num_rolls, n))
  # Calculate the sum of each pair of dice
  sums = np.sum(outcomes, axis=1)
  # Count the number of times each sum occurs
  unique_sums, counts = np.unique(sums, return_counts=True)
  # Calculate the probability density function (PDF)
  pdf = counts / num_rolls
  return unique_sums, pdf

sums, pdf = prob(num_rolls=100000, n=2)
```

```python
# Plotting the PDF for two dice
plt.figure(figsize=(6, 4))
plt.bar(sums, pdf, width=0.5, align='center', alpha=0.6)
plt.title('Probability Density Function (PDF) of Sum of Two Dice Rolls')
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.xticks(np.arange(2, 13, step=1))
plt.grid(True)
plt.tight_layout();
```

```python
sums, pdf = prob(num_rolls=100000, n=3)
# Plotting the PDF for three dice
plt.figure(figsize=(6, 4))
plt.bar(sums, pdf, width=0.5, align='center', alpha=0.6)
plt.title('Probability Density Function (PDF) of Sum of Three Dice Rolls')
plt.xlabel('Sum of three Dice')
plt.ylabel('Probability')
plt.xticks(np.arange(3, 19, step=1))
plt.grid(True)
plt.tight_layout();
```

```python
sums, pdf = prob(num_rolls=100000, n=4)
# Plotting the PDF for three dice
plt.figure(figsize=(6, 4))
plt.bar(sums, pdf, width=0.5, align='center', alpha=0.6)
plt.title('Probability Density Function (PDF) of Sum of Four Dice Rolls')
plt.xlabel('Sum of four Dice')
plt.ylabel('Probability')
plt.xticks(np.arange(4, 25, step=1))
plt.grid(True)
plt.tight_layout();
```
